```python
def generateHtml(recto):
    v_eng = ""
    v_ger = ""
    v_manuscript = ""
    v_date = ""
    r_eng = recto['eng']
    r_manuscript = recto['manuscript']
    r_ger = recto['ger']
    r_date = recto['date']
    verso = findVerso(recto)
    dot = ""
    if verso:
        v_eng = verso['eng']
        v_ger = verso['ger']
        v_manuscript = verso['manuscript']
        v_date = verso['date']
        dot = "·"
    html = """
    <div class="section">
       <div class="bead"></div>
       <div class="content verso">
          <p>{v_manuscript} {dot} {v_date}</p>
          <div class="entry">
             <p>
                <span class="eng">
                {verso_eng}
                </span>
                <span class="tlp">
                {verso_eng}
                </span>
                <span class="deu">
                {verso_ger}
                </span>
             </p>
          </div>
       </div>
       <div class="content recto">
          <p>{r_manuscript} · {r_date}</p>
          <div class="entry">
             <p>
                <span class="eng">
                {recto_eng}
                </span>
                <span class="tlp">
                {recto_eng}
                </span>
                <span class="deu">
                {recto_ger}
                </span>
             </p>
          </div>
       </div>
    </div>
    """.format(v_manuscript=v_manuscript, dot=dot, v_date=v_date, verso_eng=v_eng, verso_ger=v_ger, r_manuscript=r_manuscript,r_date=r_date,recto_eng=r_eng,recto_ger=r_ger)
    return html
    
```
